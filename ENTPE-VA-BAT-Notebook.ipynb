## 3. Optimiser la localisation des ouvrages pour éviter le recyclage

Lorsque les ouvrages de prélèvement et de rejet sont trop rapprochés, un phénomène de recyclage peut survenir (Figure 4). Ce phénomène doit être évité pour ne pas dégrader les performances de l'installations et garantir sa perennité.

Pour plus d'information : 

*Milnes, E., & Perrochet, P. (2013). Assessing the impact of thermal feedback and recycling in open-loop groundwater heat pump (GWHP) systems: a complementary design tool. Hydrogeology journal, 21(2), 505-514.*

<img src="FIG4.png" alt="Drawing1" style="width: 500px"/>
<h1 align=center><font size = 2>Fig.4 : Illustration du principe du recyclage thermique d'un dispositif géothermique sur eau de nappe (d'après Milnes et Perrochet (2013)) </font></h1>

Il existe un critère technique permettant de garantir que ce phénomène ne se produit pas. Pour cela, la distance entre le puits de pompage et de rejet doit respecter le critère suivant :

$ L > 2Q /T \pi i$

où :
* $Q$ est le débit du dispositif
* $T$ est la transimissivité de la nappe (c'est le produit de l'épaisseur productive et de la conductivité hydraulique)
* $i$ est le gradient hydraulique de la nappe.

*ATTENTION* Ce critère n'est valable que si le puits de pompage est situé à l'amont du puits d'injection, dans l'axe de l'écoulement régional.

## 4. Prendre en compte l'impact des installations existantes

Une enquête de voisinage a révélé la présence d’un dispositif géothermique à proximité du projet (voir Annexe 1). Le débit de pointe de ce dispositif est de $12 \space \text{m}^{3}/\text{h}$. 

Le script ci-dessous vous est fournit pour décrire les systèmes d'écoulement autour d'un puits de pompage ou d'un puits de rejet. Manipulez les valeurs de débit du puits, les caractéristiques physiques et la direction des écoulements pour visualiser l'effet sur la structure des systèmes d'écoulement.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

#Localisation du puits voisin
Xw = 100. #x-coordinate 
Yw = -25. #y-coordinate 
Qw = 0.01 #débit (>0 : pompage ; <0 rejet)

#Propriétés de l'aquifère et caractéristiques de l'écoulement
alpha = 15.#angle entre la direction est et la direction régionale des écoulements
h = 15. #épaisseur de l'aquifère
grad = 0.001 #gradient hydraulique
K = 0.003 #conductivité hydraulique de l'aquifère
n = 0.2 #coefficient d'emmagasinement

#Graphical Layout
Xmax = 150#x_max de la fenêtre d'affichage
Ymax = Xmax #maximal y of the window

#other useful variable
v0 = grad*K #darcy velocity (m/day)
alpha_rad = alpha*np.pi/180 #value of the angle in rad
A = Qw/K/h/grad/2

def phi(x,y):
    return -v0*(x*np.cos(alpha_rad)+y*np.sin(alpha_rad)) - (Qw/2/np.pi/h)*np.log(np.sqrt((x-Xw)**2+(y-Yw)**2))

def psi(x,y):
    return v0*(y*np.cos(alpha_rad)-x*np.sin(alpha_rad)) + (Qw/2/np.pi/h)*np.arctan2(y-Yw,x-Xw)

numcols, numrows = 100, 100
xi = np.linspace(-Xmax, Xmax, numcols)
yi = np.linspace(-Xmax, Xmax, numrows)
xi, yi = np.meshgrid(xi, yi)
head = phi(xi,yi)
stream = psi(xi,yi)

x_doi = [0,-100,-50,0,50,100]
y_doi = [-100,25,75,25,70,0]

fig, ax = plt.subplots(figsize=(8,8))
mpl.rcParams['contour.negative_linestyle'] = 'solid'

ax.grid(color='black', linestyle='-', linewidth=0.5)

im = ax.contour(xi,yi,head,25,colors='b',extent=(-Xmax,Xmax,-Ymax,Ymax),zorder = 2)
im = ax.contour(xi,yi,stream,50,colors='r',extent=(-Xmax,Xmax,-Ymax,Ymax),zorder = 2)
ax.fill(x_doi,y_doi,facecolor='lightsalmon',zorder=1)

print("Systèmes d'écoulement autour du projet:")
```

## 5. Estimer l'impact de l'installation sur la température de la nappe

Les dispositifs géothermiques sur eau de nappe génèrent un panache de température à l'aval du puits de rejet. Ce panache de température est froid lorsque l'installation fonctionne pour du chauffage, et chaud lorsque l'installation fonctionne pour du refroidissement. Il existe des solutions de calcul permettant d'estimer efficacement ces impacts thermiques. L'une d'entre elles est retenue dans le cadre de ce TD.

Pour plus d'information :

*Pophillat, W., Attard, G., Bayer, P., Hecht-Méndez, J., & Blum, P. (2018). Analytical solutions for predicting thermal plumes of groundwater heat pump systems. Renewable Energy.*

```python
#Déterminer le débit moyen Qinj

Qmoy = 0.0005 #valeur donnée à titre d'exemple
```

```python
#la fonction permettant de calculer l'impact thermique d'un rejet est définie de la façon suivante :

#LINEAR ADVECTIVE HEAT TRANSPORT MODEL
C_m = float(2888000) #volumetric heat capacity porous media (J/ kg / K)
C_w = float(4185000) #volumetric heat capacity water (J/ kg / K)
alpha_L = float(5) #longitudinal dispersivity (m)
alpha_T = float(0.5) #transverse dispersivity (m)
lambda_m = float(2.24) #thermal conductivity (W/m/K)
R = C_m/n/C_w #retardation factor
va = v0/n #seepage velocity

from scipy.special import erf,erfc
def PAHT1(x,y,Q,DTinj,alpha,t):
    alpha_rad = alpha*np.pi/180 #value of the angle in rad
    
    Y = Q/2/h/va/n
    qh = DTinj*C_w*Q
    F0 = qh/h
    DT0 = F0/va/n/C_w/Y
    Dx = lambda_m/n/C_w + alpha_L*va
    Dy = lambda_m/n/C_w + alpha_T*va
    
    x2 = x - X_rejet
    y2 = y - Y_rejet
    
    x3 = np.cos(-alpha_rad+np.pi)*x2 - np.sin(-alpha_rad+np.pi)*y2
    y3 = np.sin(-alpha_rad+np.pi)*x2 + np.cos(-alpha_rad+np.pi)*y2
    
    return DT0/4*erfc((R*x3-va*t)/(2*np.sqrt(Dx*R*t)))*(erf((y3+Y/2)/(2*np.sqrt(Dy*x3/va)))- erf((y3-Y/2)/(2*np.sqrt(Dy*x3/va))))
```

```python
DTinj = 10. #Différence de température entre le puits de pompage et le puits de rejet

t = 3600*24*30.5*6 # durée sur laquelle on souhaite simuler l'impact thermique

TEMPERATURE = PAHT1 (xi,yi,Qmoy,alpha,DTinj,t) #Calcul de l'impact thermique sur la grille

fig, ax = plt.subplots(figsize=(10,8))
mpl.rcParams['contour.negative_linestyle'] = 'solid'

ax.grid(color='black', linestyle='-', linewidth=0.5)

ax.scatter(X_pompage,Y_pompage,zorder = 3, c="black", label ="Pompage")
ax.scatter(X_rejet,Y_rejet,zorder = 3, c="red", label ="Rejet")
im = ax.contour(xi,yi,head,25,colors='b',extent=(-Xmax,Xmax,-Ymax,Ymax),zorder = 2)
im = ax.contour(xi,yi,stream,20,colors='r',extent=(-Xmax,Xmax,-Ymax,Ymax),zorder = 2)
###############################
im = ax.contourf(xi,yi,TEMPERATURE,[2,4,6,8,10],extent=(-Xmax,Xmax,-Ymax,Ymax),cmap='RdYlGn_r',zorder = 2)
###############################
ax.fill(x_doi,y_doi,facecolor='lightsalmon',zorder=1)
cb = fig.colorbar(im, ax = ax,label=u'Différence de température (K)')
ax.legend()

plt.show()
```

## 6. Déterminer un périmètre de protection autour du projet

L’article L. 134-6 du code minier prévoit que les gîtes géothermiques à basse température ne puissent être exploités qu’en vertu d’un permis d’exploitation accordé par l’autorité administrative. Ce permis d’exploitation confère un droit exclusif d’exploitation dans un volume déterminé, dit « volume d’exploitation », défini par un périmètre et deux profondeurs, conformément à l’article 14-6 de ce même code. 

La définition de ce volume d'exploitation permet de protéger l'installation de l'arrivée d'un nouveau projet géothermique susceptible d'altérer la température de l'eau prélevée.

Dans le cadre de ce TD, on considère que pour garantir les performances du projet, la température de l'eau souterraine au droit du site ne doit pas varier de plus de 2 °C. La puissance maximale exploitable autour du projet est liée à la probabilité que des transferts thermiques générés en un point donné, parviennent jusqu'au projet (cf. Figure 6).
